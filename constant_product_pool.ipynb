```python
def k(pool):
    return pool[0]*pool[1]
    
def swapA(pool, a):
    pool_value = k(pool)
    pool_a = pool[0] + a
    pool_b = pool_value / pool_a
    diff_b = pool[1] - pool_b
    pool[0] = pool_a
    pool[1] = pool_b
    return diff_b
    
    
import matplotlib.pyplot as plt  
import numpy as np  

pool = [200, 2]
prices = []
events = range(3)
for i in events:
    a = 100
    b = swapA(pool, a)
    prices.append(a/b)

plt.ylabel('B price')  
plt.xlabel('time')  
plt.grid(True)  
plt.plot(events, prices)  
plt.show()  

```
